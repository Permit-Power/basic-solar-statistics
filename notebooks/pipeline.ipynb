# Pipeline

Stage 1 of the data pipeline. Pulls raw data from EIA, LBNL, ResStock, and Solar TRACE; 
processes it into clean DataFrames; uploads each dataset to Google Drive; 
and saves all outputs as CSVs to `output_csvs/` for use by downstream notebooks.

```python
import pandas as pd
from pathlib import Path
import os
import sys
sys.path.append("../modules")

from dotenv import load_dotenv
load_dotenv("../.env")

# Modules to process data
import pull_solar_count_and_capacity as eia
from solar_technical_potential import load_solar_technical_potential
from solar_eligible_households import compute_solar_eligibility
from median_solar_costs import load_and_process_lbnl
from average_electricity_prices import build_eia861_prices
from median_permit_fees import run_pipeline as fees_runpipeline
from median_interconnection_timelines import run_pipeline as ix_runpipeline

# Module to upload data
from drive_uploader import get_drive_service, ensure_path, upload_df_to_drive

# Manifest builder
from manifest import ManifestBuilder
```

```python
# To add all CSVs to output

OUTPUT_DIR = "../output_csvs"
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
# Configurations to upload data to Google Drive

# Identify folder ID (the Big Numbers Database)
ROOT_ID = "1DBlVUvspIPTTyZPtVovYtEgUSmQNXmG7"

service = get_drive_service()

# Start collecting metadata for the manifest
mb = ManifestBuilder()
```

```python
# Load data

# ResStock metadata for solar potential calcs
resstock_metadata = pd.read_csv("../data/resstock_metadata_technical_potential.csv")

# AHJ population data
ahj_pop = pd.read_csv("../data/ahj_distribution_permitting_timelines_pv.csv")[['state', 'name', 'geoid', 'population']]

# Solar TRACE permitting fees
fees = pd.read_csv("../data/solartrace_fees.csv")

# Solar TRACE interconnection timelines
ix = pd.read_csv("../data/solartrace_ix.csv")

# State name <> abbr mapping
name_abbr = pd.read_csv("../data/state_name_abbr.csv")
```

```python
# Certain datasets were processed manually based on Ohm data, we will simply load those and upload them without any processing

# AHJ
cancellation_ahj = pd.read_csv("../data/ahj_cancellation_rates.csv")
permitting_timelines_dist_ahj = pd.read_csv("../data/ahj_distribution_permitting_timelines_pv.csv")
permitting_timelines_tech_ahj = pd.read_csv("../data/ahj_median_permitting_timelines_by_tech.csv").drop(columns=['source_id'])
inspection_timelines_ahj = pd.read_csv("../data/ahj_inspection_timelines_pv.csv").drop(columns=['source_id'])

# State
cancellation_state = pd.read_csv("../data/state_cancellation_rates.csv")
permitting_timelines_dist_state = pd.read_csv("../data/state_distribution_permitting_timelines_pv.csv")
permitting_timelines_tech_state = pd.read_csv("../data/state_median_permitting_timelines_by_tech.csv")
inspection_timelines_state = pd.read_csv("../data/state_inspection_timelines_pv.csv")
solar_potential_twh = load_solar_technical_potential()
solar_jobs = pd.read_csv("../data/solar_jobs.csv")


# Upload to Google Drive
folder_id = ensure_path(service, ROOT_ID, ["1. Solar"])

fid = upload_df_to_drive(solar_potential_twh, folder_id, "Potential Solar Generation by State")
mb.register(solar_potential_twh, fid, "1. Solar",
            "Potential Solar Generation by State",
            description="Technical solar generation potential by state in terawatt-hours.",
            source="ResStock / NREL", granularity="state")

folder_id = ensure_path(service, ROOT_ID, ["2. Permitting"])

fid = upload_df_to_drive(cancellation_ahj, folder_id, "Cancellation Rates by AHJ")
mb.register(cancellation_ahj, fid, "2. Permitting",
            "Cancellation Rates by AHJ",
            description="Percentage of solar permit applications cancelled, by jurisdiction.",
            source="Ohm Analytics", granularity="ahj")

fid = upload_df_to_drive(cancellation_state, folder_id, "Cancellation Rates by State")
mb.register(cancellation_state, fid, "2. Permitting",
            "Cancellation Rates by State",
            description="Percentage of solar permit applications cancelled, by state.",
            source="Ohm Analytics", granularity="state")

fid = upload_df_to_drive(permitting_timelines_dist_ahj, folder_id, "Distribution of Solar Permitting Timelines by AHJ")
mb.register(permitting_timelines_dist_ahj, fid, "2. Permitting",
            "Distribution of Solar Permitting Timelines by AHJ",
            description="Full distribution (percentiles) of permitting timelines in days, by jurisdiction.",
            source="Ohm Analytics", granularity="ahj")

fid = upload_df_to_drive(permitting_timelines_dist_state, folder_id, "Distribution of Solar Permitting Timelines by State")
mb.register(permitting_timelines_dist_state, fid, "2. Permitting",
            "Distribution of Solar Permitting Timelines by State",
            description="Full distribution (percentiles) of permitting timelines in days, by state.",
            source="Ohm Analytics", granularity="state")

fid = upload_df_to_drive(permitting_timelines_tech_ahj, folder_id, "Median Permitting Timeline by Tech by AHJ")
mb.register(permitting_timelines_tech_ahj, fid, "2. Permitting",
            "Median Permitting Timeline by Tech by AHJ",
            description="Median permitting timeline in days, broken out by technology (solar vs storage), by jurisdiction and year.",
            source="Ohm Analytics", granularity="ahj")

fid = upload_df_to_drive(permitting_timelines_tech_state, folder_id, "Median Permitting Timeline by Tech by State")
mb.register(permitting_timelines_tech_state, fid, "2. Permitting",
            "Median Permitting Timeline by Tech by State",
            description="Median permitting timeline in days, broken out by technology (solar vs storage), by state and year.",
            source="Ohm Analytics", granularity="state")

folder_id = ensure_path(service, ROOT_ID, ["3. Inspection"])

fid = upload_df_to_drive(inspection_timelines_ahj, folder_id, "Median Inspection Timeline by AHJ")
mb.register(inspection_timelines_ahj, fid, "3. Inspection",
            "Median Inspection Timeline by AHJ",
            description="Median inspection timeline in days, by jurisdiction.",
            source="Ohm Analytics", granularity="ahj")

fid = upload_df_to_drive(inspection_timelines_state, folder_id, "Median Inspection Timeline by State")
mb.register(inspection_timelines_state, fid, "3. Inspection",
            "Median Inspection Timeline by State",
            description="Median inspection timeline in days, by state.",
            source="Ohm Analytics", granularity="state")

folder_id = ensure_path(service, ROOT_ID, ["6. Jobs"])

fid = upload_df_to_drive(solar_jobs, folder_id, "Solar Jobs by State by Category")
mb.register(solar_jobs, fid, "6. Jobs",
            "Solar Jobs by State by Category",
            description="Solar employment counts by state, broken down by job category (installation, manufacturing, etc.).",
            source="Solar Foundation / Interstate Renewable Energy Council", granularity="state")
```

```python
# Solar/Storage capacity and installations by year by sector

years = range(2017, 2026)

capacity = eia.download_and_aggregate_distributed_solar(years)

# Upload to drive
folder_id = ensure_path(service, ROOT_ID, ["1. Solar"])

fid = upload_df_to_drive(capacity, folder_id, "Solar/Storage Capacity and Installations by State and Sector Over Time")
mb.register(capacity, fid, "1. Solar",
            "Solar/Storage Capacity and Installations by State and Sector Over Time",
            description="Cumulative solar and storage capacity (MW) and installation counts by state, sector, and year.",
            source="EIA 861M", granularity="state")
```

```python
# Solar eligibility

eligibility = compute_solar_eligibility(resstock_metadata)

# Upload to drive
folder_id = ensure_path(service, ROOT_ID, ["1. Solar"])

fid = upload_df_to_drive(eligibility, folder_id, "Count and Percentage of Solar-eligible Households by State")
mb.register(eligibility, fid, "1. Solar",
            "Count and Percentage of Solar-eligible Households by State",
            description="Estimated number and percentage of households eligible for rooftop solar, by state.",
            source="ResStock / NREL", granularity="state")
```

```python
# Median solar costs over time

import os
from dotenv import load_dotenv
load_dotenv("../.env")

# Google Drive link to the LBNL Tracking the Sun ZIP file
drive_link = "https://drive.google.com/file/d/1NQh4TRC_IqDz2r5vfZuxDm6LGjEuexdu/view"

# FRED API Key (loaded from .env locally, or from GitHub Actions secret in CI)
fred_api_key = os.environ["FRED_API_KEY"]

costs = load_and_process_lbnl(
    drive_url=drive_link,
    fred_api_key=fred_api_key,
    start_year=2000,        # CPI start year
    min_install_year=2000,  # Only installations from 2000 onward
)

# Upload to drive
folder_id = ensure_path(service, ROOT_ID, ["1. Solar"])

fid = upload_df_to_drive(costs, folder_id, "Residential Solar Costs by State Over Time")
mb.register(costs, fid, "1. Solar",
            "Residential Solar Costs by State Over Time",
            description="Median residential solar installation cost per kW (inflation-adjusted) and median system size by state and year.",
            source="LBNL Tracking the Sun", granularity="state")
```

```python
# Average electricity prices

years = range(2015, 2025)

util_prices, state_prices = build_eia861_prices(years)

# Upload to drive
folder_id = ensure_path(service, ROOT_ID, ["5. Rates"])

fid = upload_df_to_drive(state_prices, folder_id, "Average Electricity Rates by State")
mb.register(state_prices, fid, "5. Rates",
            "Average Electricity Rates by State",
            description="Average electricity price per kWh by state, sector, and year.",
            source="EIA 861", granularity="state")

fid = upload_df_to_drive(util_prices, folder_id, "Average Electricity Rates by Utility")
mb.register(util_prices, fid, "5. Rates",
            "Average Electricity Rates by Utility",
            description="Average electricity price per kWh by utility, sector, and year.",
            source="EIA 861", granularity="utility")
```

```python
# Side bar to extract top utilities for interconnection fee research

# (1) Only residential
util_prices = util_prices[util_prices["sector"] == "Residential"]

#  (2) Top 100 utilities nationwide
top100 = util_prices.nlargest(100, "customers")

#  (3) Top 2 utilities per state
top2_per_state = (
    util_prices
    .groupby("state", group_keys=False)
    .head(2)          # because the df is already sorted by customers descending
)

# (4) Combine both conditions, remove duplicates
filtered = pd.concat([top100, top2_per_state]).drop_duplicates()

# (5) Sort nicely
filtered = filtered.sort_values(["state", "customers"], ascending=[True, False])

# Write locally
# filtered[['utility_name', 'utility_number', 'state', 'customers']].to_csv("../data/utility_interconnection_fees.csv", index=False)
```

```python
# Permitting fees

ahj_df_fees, state_df_fees = fees_runpipeline(fees, ahj_pop, name_abbr)

# Upload to drive
folder_id = ensure_path(service, ROOT_ID, ["2. Permitting"])

fid = upload_df_to_drive(state_df_fees, folder_id, "Permitting Fees by State")
mb.register(state_df_fees, fid, "2. Permitting",
            "Permitting Fees by State",
            description="Median solar permitting fee in dollars, by state.",
            source="Solar TRACE", granularity="state")

fid = upload_df_to_drive(ahj_df_fees, folder_id, "Permitting Fees by AHJ")
mb.register(ahj_df_fees, fid, "2. Permitting",
            "Permitting Fees by AHJ",
            description="Median solar permitting fee in dollars, by jurisdiction.",
            source="Solar TRACE", granularity="ahj")
```

```python
# Interconnection timelines

ahj_df_ix, state_df_ix = ix_runpipeline(ix)

# Upload to drive
folder_id = ensure_path(service, ROOT_ID, ["4. Interconnection"])

fid = upload_df_to_drive(state_df_ix, folder_id, "Interconnection Timelines by State")
mb.register(state_df_ix, fid, "4. Interconnection",
            "Interconnection Timelines by State",
            description="Median interconnection timelines (pre-install, final IX to PTO) in days, by state and year.",
            source="Solar TRACE", granularity="state")

fid = upload_df_to_drive(ahj_df_ix, folder_id, "Interconnection Timelines by AHJ")
mb.register(ahj_df_ix, fid, "4. Interconnection",
            "Interconnection Timelines by AHJ",
            description="Median interconnection timelines (pre-install, final IX to PTO) in days, by jurisdiction.",
            source="Solar TRACE", granularity="ahj")
```

```python
# Average and lifetime bill savings with solar (from exported CSV)

from solar_bill_savings_update import load_from_export

savings = load_from_export("../data/bill_savings_csvs")
savings = savings[["state_abbr", "year_1_savings", "lifetime_savings"]]

# Upload to Google Drive
folder_id = ensure_path(service, ROOT_ID, ["1. Solar"])
fid = upload_df_to_drive(savings, folder_id, "Average Annual and Lifetime Solar Savings by State")
mb.register(savings, fid, "1. Solar",
            "Average Annual and Lifetime Solar Savings by State",
            description="Weighted median annual and lifetime bill savings from residential solar, by state.",
            source="dGen model output", granularity="state")
```

```python
# ── Export all processed DataFrames as CSVs ─────────────────────────


# Solar
solar_potential_twh.to_csv(f"{OUTPUT_DIR}/potential_solar_generation_by_state.csv", index=False)
capacity.to_csv(f"{OUTPUT_DIR}/solar_storage_capacity_installations_by_state_sector.csv", index=False)
eligibility.to_csv(f"{OUTPUT_DIR}/solar_eligible_households_by_state.csv", index=False)
costs.to_csv(f"{OUTPUT_DIR}/residential_solar_costs_by_state_over_time.csv", index=False)
savings.to_csv(f"{OUTPUT_DIR}/annual_and_lifetime_solar_savings_by_state.csv", index=False)
 
# Permitting
cancellation_state.to_csv(f"{OUTPUT_DIR}/cancellation_rates_by_state.csv", index=False)
cancellation_ahj.to_csv(f"{OUTPUT_DIR}/cancellation_rates_by_ahj.csv", index=False)
permitting_timelines_dist_state.to_csv(f"{OUTPUT_DIR}/permitting_timeline_distribution_by_state.csv", index=False)
permitting_timelines_dist_ahj.to_csv(f"{OUTPUT_DIR}/permitting_timeline_distribution_by_ahj.csv", index=False)
permitting_timelines_tech_state.to_csv(f"{OUTPUT_DIR}/median_permitting_timeline_by_tech_by_state.csv", index=False)
permitting_timelines_tech_ahj.to_csv(f"{OUTPUT_DIR}/median_permitting_timeline_by_tech_by_ahj.csv", index=False)
state_df_fees.to_csv(f"{OUTPUT_DIR}/permitting_fees_by_state.csv", index=False)
ahj_df_fees.to_csv(f"{OUTPUT_DIR}/permitting_fees_by_ahj.csv", index=False)
 
# Inspection
inspection_timelines_state.to_csv(f"{OUTPUT_DIR}/inspection_timelines_by_state.csv", index=False)
inspection_timelines_ahj.to_csv(f"{OUTPUT_DIR}/inspection_timelines_by_ahj.csv", index=False)
 
# Interconnection
state_df_ix.to_csv(f"{OUTPUT_DIR}/interconnection_timelines_by_state.csv", index=False)
ahj_df_ix.to_csv(f"{OUTPUT_DIR}/interconnection_timelines_by_ahj.csv", index=False)
 
# Rates
state_prices.to_csv(f"{OUTPUT_DIR}/electricity_rates_by_state.csv", index=False)
util_prices.to_csv(f"{OUTPUT_DIR}/electricity_rates_by_utility.csv", index=False)
 
# Jobs
solar_jobs.to_csv(f"{OUTPUT_DIR}/solar_jobs_by_state.csv", index=False)
 
print(f"Exported {len(os.listdir(OUTPUT_DIR))} CSV files to {OUTPUT_DIR}/")
```

```python
# Generate and upload manifest

# Save locally as reference
mb.save("../manifest.json")

# Upload to root of Big Numbers Database on Google Drive
mb.upload_to_drive(service, ROOT_ID)
```
